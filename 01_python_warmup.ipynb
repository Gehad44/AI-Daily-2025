```python
# ===== NumPy =====
import numpy as np
arr = np.random.randint(10, size=7)
print("Array:", arr)
print("Mean:", arr.mean(), " | Std:", arr.std())

# ===== Pandas =====
import pandas as pd
df = pd.DataFrame({
    "age": np.random.randint(20, 40, 10),
    "salary": np.random.randint(3000, 8000, 10)
})
print("\nFirst 5 rows:")
print(df.head())
```
